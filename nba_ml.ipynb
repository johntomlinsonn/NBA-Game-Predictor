```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv("nba.csv")
```

```python
def clean_csv(precsv):
    rows_to_drop = []

    for i in range(len(precsv)):
        if np.isnan(precsv.at[i, "season-avgOPP-FT%"]) or np.isnan(precsv.at[i, "avg-FGA"]):
            rows_to_drop.append(i)

    precsv.drop(rows_to_drop, inplace=True)
    return precsv

games = clean_csv(games)
```

```python
games
```

```python
def data_types_tostring():
    data_types = games.dtypes.tolist()
    columns = games.columns.tolist()
    for i in range(len(data_types)):
        print(str(columns[i]) + ": " + str(data_types[i]))
```

```python
data_types_tostring()
```

```python
def clean_ot(precsv):
    ot = precsv["OT"].to_numpy()
    for i in range(len(ot)):
        if ot[i] == "OT":
            ot[i] = 1
        else:
            ot[i] = 0
    precsv["OT"] = ot
    return precsv

clean_ot(games)
data_types_tostring()

```

```python
games["Date"] = pd.to_datetime(games["Date"])
games['DayOfWeek'] = games['Date'].dt.dayofweek
games['Month'] = games['Date'].dt.month
games['Season_Progress'] = games['Date'].dt.dayofyear % 365 / 365
```

```python
team_win_rates = games.groupby('Team')['Won'].rolling(window=10).mean().reset_index(0, drop=True).fillna(0.5)
games['Team_Win_Rate_Last10'] = games['Team'].map(team_win_rates)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import precision_score, accuracy_score, f1_score, roc_auc_score, precision_recall_curve
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.feature_selection import SelectFromModel
import seaborn as sns
import time

def improve_xgboost_model(X_train, y_train, X_test, y_test, predictors):
    """
    Comprehensive approach to improve XGBoost model performance
    """
    results = {}
    start_time = time.time()
    
    print("Starting model improvement process...")
    
    # 1. Feature Importance Analysis from Base Model
    print("\nStep 1: Feature Importance Analysis")
    base_model = xgb.XGBClassifier(random_state=42)
    base_model.fit(X_train, y_train)
    
    # Get feature importances
    feature_importances = pd.DataFrame({
        'feature': predictors,
        'importance': base_model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print(f"Top 10 important features:\n{feature_importances.head(10)}")
    
    # Plot top features
    plt.figure(figsize=(10, 6))
    sns.barplot(x='importance', y='feature', data=feature_importances.head(20))
    plt.title('Top 20 Feature Importances')
    plt.tight_layout()
    plt.show()
    
    # 2. Feature Selection - Use top 80% of cumulative importance
    print("\nStep 2: Feature Selection")
    feature_importances['cumulative_importance'] = feature_importances['importance'].cumsum() / feature_importances['importance'].sum()
    selected_features = feature_importances[feature_importances['cumulative_importance'] <= 0.80]['feature'].tolist()
    
    print(f"Selected {len(selected_features)} features out of {len(predictors)}")
    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]
    
    # 3. Hyperparameter Tuning with Cross-Validation
    print("\nStep 3: Hyperparameter Tuning")
    param_grid = {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7],
        'subsample': [0.7, 0.8, 0.9],
        'colsample_bytree': [0.7, 0.8, 0.9],
        'gamma': [0, 0.1, 0.2]
    }
    
    # Define cross-validation strategy
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    # Create grid search
    grid_search = GridSearchCV(
        estimator=xgb.XGBClassifier(random_state=42),
        param_grid=param_grid,
        cv=cv,
        scoring='precision',
        n_jobs=-1,
        verbose=1
    )
    
    # Fit grid search
    grid_search.fit(X_train_selected, y_train)
    
    best_params = grid_search.best_params_
    print(f"Best parameters: {best_params}")
    
    # 4. Train Final Model with Best Parameters
    print("\nStep 4: Training Final Model with Best Parameters")
    final_model = xgb.XGBClassifier(
        n_estimators=best_params['n_estimators'],
        learning_rate=best_params['learning_rate'],
        max_depth=best_params['max_depth'],
        subsample=best_params['subsample'],
        colsample_bytree=best_params['colsample_bytree'],
        gamma=best_params['gamma'],
        random_state=42
    )
    
    final_model.fit(X_train_selected, y_train)
    
    # 5. Threshold Optimization
    print("\nStep 5: Threshold Optimization")
    y_prob = final_model.predict_proba(X_test_selected)[:, 1]
    
    # Calculate precision at various thresholds
    precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
    
    # Plot precision-recall curve
    plt.figure(figsize=(10, 6))
    plt.plot(recall, precision, marker='.')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.grid(True)
    plt.show()
    
    # Calculate F1 score at different thresholds
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
    optimal_idx = np.argmax(f1_scores)
    optimal_threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else 0.5
    
    print(f"Optimal threshold: {optimal_threshold:.4f}")
    
    # 6. Evaluate Model with Optimal Threshold
    print("\nStep 6: Evaluating Final Model")
    # Using optimal threshold
    y_pred_optimal = (y_prob >= optimal_threshold).astype(int)
    
    # Calculate metrics
    results['precision'] = precision_score(y_test, y_pred_optimal)
    results['accuracy'] = accuracy_score(y_test, y_pred_optimal)
    results['f1'] = f1_score(y_test, y_pred_optimal)
    results['roc_auc'] = roc_auc_score(y_test, y_prob)
    
    print("Final Model Performance with Optimal Threshold:")
    print(f"Precision: {results['precision']:.4f}")
    print(f"Accuracy: {results['accuracy']:.4f}")
    print(f"F1 Score: {results['f1']:.4f}")
    print(f"ROC AUC: {results['roc_auc']:.4f}")
    
    # 7. Compare with Default Threshold (0.5)
    y_pred_default = (y_prob >= 0.5).astype(int)
    
    print("\nDefault Threshold (0.5) Performance:")
    print(f"Precision: {precision_score(y_test, y_pred_default):.4f}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred_default):.4f}")
    print(f"F1 Score: {f1_score(y_test, y_pred_default):.4f}")
    
    # Report execution time
    end_time = time.time()
    print(f"\nModel improvement completed in {(end_time - start_time)/60:.2f} minutes")
    
    # Return both the model and optimal threshold
    return {
        'model': final_model,
        'optimal_threshold': optimal_threshold,
        'selected_features': selected_features,
        'best_params': best_params,
        'metrics': results
    }
```

```python
improved_results = improve_xgboost_model(
    X_train=train[predictors], 
    y_train=train["Won"],
    X_test=test[predictors],
    y_test=test["Won"],
    predictors=predictors
)

# To make predictions with optimal threshold:
y_prob = improved_results['model'].predict_proba(test[improved_results['selected_features']])[:, 1]
y_pred = (y_prob >= improved_results['optimal_threshold']).astype(int)

# Save the improved model and its components
import joblib
joblib.dump(improved_results, "improved_nba_model_package.pkl")
```
